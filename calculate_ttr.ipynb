```python
from collections import Counter

def calculate_ttr(text):
    # Tokenize the text by spaces (this is a very basic tokenizer)
    tokens = text.split()
    # Count unique words
    types = len(set(tokens))
    # Count total words
    total_tokens = len(tokens)
    # Calculate Type-Token Ratio
    ttr = types / total_tokens if total_tokens > 0 else 0
    return ttr

# Example text
text = "The cat sat on the mat."
ttr_value = calculate_ttr(text)
ttr_value

def calculate_ttr_corrected(text):
    # Tokenize the text by spaces (this is a very basic tokenizer)
    tokens = text.lower().split()  # Convert to lower case for consistent counting
    # Count unique words
    types = len(set(tokens))
    # Count total words
    total_tokens = len(tokens)
    # Calculate Type-Token Ratio
    ttr = types / total_tokens if total_tokens > 0 else 0
    return ttr

# Re-run with corrected case sensitivity
corrected_ttr_value = calculate_ttr_corrected(text)
corrected_ttr_value
```
